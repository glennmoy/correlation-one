```python
from math import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
df=pd.read_csv('geographic.csv')

neighbourhoods=df.columns
```

```python
def get_centroid(bound):
    cen_x=0.5*(max(bound[:,0]+min(bound[:,0])))
    cen_y=0.5*(max(bound[:,1]+min(bound[:,1])))
    return [cen_x,cen_y]
```

```python
# Find the neighbourhood of a point in lat,long

def get_nearest_neighbour(point):
    
    min_dist=10000
    
    for nbr in df_centroid:
        x=df_centroid[nbr]['cent_x']
        y=df_centroid[nbr]['cent_y']
        
        dist=np.sqrt((point[0]-x)**2+(point[1]-y)**2)
        
        if dist < min_dist:
            min_dist=dist
            point_neighbourhood=nbr
    
    return point_neighbourhood,dist,[x,y]
```

```python
# Make new dataframe with x,y values
df_centroid={}

# Find all the centroids of the neighbourhoods
for nbrhood in df:
    boundary=df[nbrhood].as_matrix().reshape(int(df[nbrhood].size/2),2)
    cent=get_centroid(boundary)
    
    case={'cent_x':cent[0],'cent_y':cent[1]}
    df_centroid[nbrhood]=case
    
    plt.plot(boundary[:,0],boundary[:,1])
    plt.scatter(cent[0],cent[1],color='r',s=0,)

plt.show()

df_centroid=df.from_dict(df_centroid)
```
